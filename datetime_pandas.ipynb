```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
index = pd.date_range('2021', freq='D', periods=365)
values = np.random.normal(10, 2, len(index))

ts = pd.Series(values, index=index)
```

```python
ts.plot()
```

```python
idx = pd.date_range('2019', freq='H', periods=10)
ts_hours = pd.Series(np.random.normal(10, 2, len(idx)), index=idx)

ts_hours
```

```python
ts_hours.plot()
```

```python
resampled = ts_hours.resample('2H').mean()

resampled.plot()
```

```python
interpolated = ts_hours.resample('0.5H').interpolate()

interpolated.plot()
```
